## Competition Functions

```python
def inoculateOvernightCulture(strain, media, media_volume=5.0, name=""):
    overnight = CultureVolume(name="Overnight "+name)
    overnight.add(media.sample(media_volume))
    overnight.add(freezedowns[strain.name].sample(0.010))
    return overnight
```

```python
def inoculateDilutedBackCulture(strain, media, overnight, media_volume=25.0, overnight_volume=250.0, name=""):
    dilutedBack = CultureVolume(name="Diluted-back "+name)
    dilutedBack.add(media.sample(media_volume))
    dilutedBack.add(tetStock.sample(media_volume/1000))
    dilutedBack.add(overnight.sample(overnight_volume))
    return dilutedBack
```

```python
def competitionFlaskTransfer(competition_flask, media, media_volume, transfer_volume, name=""):
    competitionFlask_next = CultureVolume(name="Competition "+name)
    competitionFlask_next.add(media.sample(competitionFlaskMediaVolume))
    competitionFlask_next.add(tetStock.sample(competitionFlaskMediaVolume/1000))
    competitionFlask_next.add(competition_flask.sample(transfer_volume))
    return competitionFlask_next
```

```python
def collectEnumerationSampleData(competition_culture, sample_reps, t, transfer_volumes, data, sim_flow_cytometry=True):
    for r in range(sample_reps):
        enumerationSample         = competition_culture.sample(1.0)
        # enumerationSample.in fo()
        enumerationDilutionSeries = DilutionSeries(enumerationSample)
        
        flowReadVolume    = 0.100
        flowReadDilutions = [-2, -1]
        for df in flowReadDilutions:
            
            if(sim_flow_cytometry):
                flowEvents = flow.read(sample=enumerationDilutionSeries.getDilution(df), read_volume=flowReadVolume)
                totalDensityEstimate = 0.0
                totalPopSizeEstimate = 0.0
                for markerName, numEvents in flowEvents.iteritems():
                    markerDensityEstimate = numEvents*(1.0/flowReadVolume) * 10**(numpy.abs(df)) 
                    markerPopSizeEstimate = markerDensityEstimate * reduce((lambda x, y: x * y), [competitionFlaskMediaVolume/xfrvol for xfrvol in transfer_volumes], 1)  # (competitionFlaskMediaVolume/competitionFlaskDefaultTransferVolume)**transferCount
                    totalDensityEstimate += markerDensityEstimate
                    totalPopSizeEstimate += markerPopSizeEstimate
                    data['pop_sizes'].append( {'time': t, 'sample_rep': r+1, 'dilution': df,
                                               'population': markerName, 'pop_size': markerPopSizeEstimate, 'density':markerDensityEstimate} )
                data['pop_sizes'].append( {'time': t, 'sample_rep': r+1, 'dilution': df,
                                           'population': 'total', 'pop_size': totalPopSizeEstimate, 'density':totalDensityEstimate} )
           
            else:
                totalDensityEstimate = 0.0
                totalPopSizeEstimate = 0.0
                for markerName, markerDensity in enumerationDilutionSeries.getDilution(df).getMarkerDensities().iteritems():
                    markerDensityEstimate = markerDensity * 10**(numpy.abs(df))
                    markerPopSizeEstimate = markerDensityEstimate * reduce((lambda x, y: x * y), [competitionFlaskMediaVolume/xfrvol for xfrvol in transfer_volumes], 1)  # (competitionFlaskMediaVolume/competitionFlaskDefaultTransferVolume)**transferCount
                    totalDensityEstimate += markerDensityEstimate
                    totalPopSizeEstimate += markerDensityEstimate
                    data['pop_sizes'].append( {'time': t, 'sample_rep': r+1, 'dilution': df,
                                               'population': markerName, 'pop_size': markerPopSizeEstimate, 'density':markerDensityEstimate} )
                data['pop_sizes'].append( {'time': t, 'sample_rep': r+1, 'dilution': df,
                                           'population': 'total', 'pop_size': totalPopSizeEstimate, 'density':totalDensityEstimate} )
```

```python
def collectSequencingSampleData(competition_culture, sample_reps, t, data, sequencing_error=0.01, frequency_precision=0.001):
    for r in range(sample_reps):
        sequencingSample         = competition_culture.sample( competition_culture.totalVolume() )
        
        # In lieu of simulating sequencing more directly, return estimated frequencies based on cell counts with small noise:
        trueCellCounts = sequencingSample.getCellCounts()
        estimatedCellCounts = {}
        for inoculumName, cellCount in trueCellCounts.iteritems():
            estimatedCellCounts[inoculumName] = int(numpy.random.normal(cellCount, cellCount*sequencing_error))
        estimatedTotalCellCount = numpy.sum(estimatedCellCounts.values())

        for inoculumName, cellCount in estimatedCellCounts.iteritems():
            frequencyEstimate = round( cellCount/estimatedTotalCellCount, int(numpy.log10(1/frequency_precision)) )
            data['allele_freqs'].append( {'time': t, 'sample_rep': r+1, 
                                          'allele': inoculumName, 'allele_freq': frequencyEstimate} ) 
```

## Start overnight cultures
* Pipette 200 mL of LB media into the overnight media prep reservoir.
* Pipette 5 mL of LB only media into the overnight neg ctrl tube
* Pipette 195 uL of Tetracycline stock into the overnight media prep reservoir.
* Pipette 5.0 mL of LB+Tet media into the 1X overnight culture tube for each strain
* Pipette 4.95 mL of LB+Tet media into the 0.1X, 0.01X, 0.001X overnight tubes for each strain

For each of the the competition strains:
    * Insert freezedown culture scraping into the corresponding 1X overnight culture tubes
    * Pipette 50 uL from the 1X overnight tube to the 0.1X overnight tube
    * Pipette 50 uL from the 0.1X overnight tube to the 0.01X overnight tube
    * Pipette 50 uL from the 0.01X overnight tube to the 0.001X overnight tube
    * Place all overnight culture tubes in shaking incubator

```python
overnights = {}

overnightMediaPrepReservoir = CultureVolume(name="Overnight Media Prep Reservoir")
overnightMediaPrepReservoir.add(LBStock.sample(250.0))

overnightMediaPrepReservoir.add(tetStock.sample(0.250))

numOvernightsPerStrain = 1
for strainName, strain in strains.items():
    overnights[strainName] = []
    for overnightRep in range(numOvernightsPerStrain):
        overnightName = strainName+"r"+str(overnightRep)
        overnights[strainName].append( inoculateOvernightCulture(strain=strain, 
                                                                 media=overnightMediaPrepReservoir, 
                                                                 media_volume=5.0, 
                                                                 name=overnightName) )

incubator.incubate(cultures=numpy.concatenate(overnights.values()), time=24, dt=0.01, record_growth_curves=False)

for overnight in numpy.concatenate(overnights.values()):
    overnight.info()
overnightMediaPrepReservoir.info()
```

## Dilute back each overnight culture:
#### Prep diluted-back culture flasks with media:
* Retrieve the 1 L flask of LB+Tet media.
* For each of the following diluted-back cultures:
    * rpobWT_bc11GFP (5 culture flasks)
    * rpobM2_bc9RFP (1 culture flasks)
    * rpobM3_bc2RFP (3 culture flasks)
    * rpobM4_bc8RFP (1 culture flasks)
    * rpobM6_bc10RFP (1 culture flasks)
    * rpobM8_bc12RFP (1 culture flasks)
    * rpobM9_bc3RFP (3 culture flasks)
    * rpobM10_bc13RFP (3 culture flasks)
* Glass Pipette 25 mL of LB+Tet media from the 1 L flask into each diluted-back culture flask.
* Put the 1 L flask of LB+Tet media to the standing incubator (for contamination monitoring).
* Put diluted-back culture flasks aside in the dark until the next step.

#### Dilute back each overnight culture:
 * Take overnight cultures of of the incubator.
 * Pipette 250 uL of each overnight culture into the corresponding diluted-back culture flask(s).
 * Place the diluted-back culture flasks in shaking incubator.
 * Note starting time of diluted-back culture growth. Prepare to take diluted-back cultures out of incubator exactly 1.0 hours later for immediate use.

```python
dilutedBacks = {}

dilutedBackMediaVolume     = 25.0
dilutedBackOvernightVolume = 0.250
numDilutedBacksPerStrain   = 5
for strainName, strain in strains.items():
    dilutedBacks[strainName] = []
    for dbRep in range(numDilutedBacksPerStrain):
        dilutedBackName = strainName
        overnight = [on for on in overnights[strainName] if on.totalVolume()>dilutedBackOvernightVolume][0]

        dilutedBacks[strainName].append( inoculateDilutedBackCulture(strain, media=LBStock, overnight=overnight, 
                                                                        media_volume=dilutedBackMediaVolume, 
                                                                        overnight_volume=dilutedBackOvernightVolume, 
                                                                        name=dilutedBackName+' r'+str(dbRep)) )

dilutedBackGrowthTime = 1.0
allDilutedBacks = [db for strainDilutedBackReps in dilutedBacks.values() for db in strainDilutedBackReps]
dbGrowthCurves = incubator.incubate(cultures=allDilutedBacks, time=dilutedBackGrowthTime, dt=0.01, record_growth_curves=True)
dbGrowthCurves['time'] -= dilutedBackGrowthTime  # label the diluted-back time series with negative time values leading up to 0:00

for dilutedBack in allDilutedBacks:
    dilutedBack.info()   
```

## Initiate Competitions

#### Inoculate competition flasks
For each strains:
* Pour the contents of all of the strain’s diluted back culture flask(s) (some strains have 1, some more than 1) into that strain’s combined diluted-back culture reservoir.

For each competition:
1. Transfer Culture from 0h Conical Tube to 4h Competition Flask:
    * Pipette 1000 uL of the culture from the 0h conical tube to the 4h competition flask.
        * (For the WTctrl 4h competition flask, pipette 1000 uL of the culture from the diluted-back culture reservoir to the 4h competition flask.)

2. Sample Culture for Flow Cytometry:
    * Pipette 1 mL of culture from the 4h competition flask into the corresponding 0h flow sample epi. 

3. Incubate Competition Flasks:
    * Place 4h competition flask in 37C shaking incubator

4. Sample Culture for Sequencing:
    * Leave the remaining 0h conical tube contents (~24 mL) in what is now the 0h sequencing sample conical tube.

```python
combinedDilbutedBacks = {}

for strainName, dilutedBacks in dilutedBacks.iteritems():
    combinedDilbutedBacks[strainName] = CultureVolume(name="Combined Diluted-back Reservoir "+strainName)
    for dilutedBackRep in dilutedBacks:
        combinedDilbutedBacks[strainName].add(dilutedBackRep.sample(dilutedBackRep.totalVolume()))

for combinedDilbutedBack in combinedDilbutedBacks.values():
    combinedDilbutedBack.info() 


competitionFlasks = {}

for competition, initStrainFreqs in competitionInitStrainFreqs.items():
    competitionFlasks[competition] = CultureVolume(name=competition)
    for strainName, initFreq in initStrainFreqs.items():
        competitionFlasks[competition].add(combinedDilbutedBacks[strainName].sample(initFreq*competitionFlaskMediaVolume))
```

### Execute competition transfers and data collection sampling

#### Fill the Competition Flasks for the next transfer:
* Pipette 25 mL of LB+Tet media into the next competition flask.

#### Perform a Spot-Check Flow Read to gauge population growth in the current interval:
* 30 mins prior to the transfer, sample the density of the WTctrl and 1-2 other competitions.
* Use this spot check density information to project competition densities at the upcoming and following transfer times and calculate the transfer volumes that will keep these projected densities in the desired range (protocol executed by excel spreadsheet).

#### Transfer Competition Culture to next Competition Flask:
* Pipette [calculated transfer volume] of the culture from the current competition flask to the next competition flask.

#### Incubate Competition Flasks:
* Place next competition flask in 37C shaking incubator

#### Sample Culture for Flow Cytometry:
* Pipette 1 mL of culture from the current competition flask into the corresponding flow sample epi for this time point. 

#### Sample Culture for Sequencing:
* Pour the remaining culture contents (~24 mL) of the current competition flask into the corresponding sequencing sample conical tube for this time point.

```python
t = 0

transferCount = 0
transferVolumes = {competition : [] for competition, flask in competitionFlasks.items()}

while t <= totalTime:
    print "\t\tT=" + str(t) + "h"

    transferVolume_adjusted = competitionFlaskDefaultTransferVolume

    for competition, competitionFlask in sorted(competitionFlasks.items(), key=lambda flask: flask[-1].totalCellDensity(), reverse=False):
        print "\t\t\t" + competition

        #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
        # Transfer Competition Culture to next Competition Flask:
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
        #.............................
        # Dynamically calculate the appropriate transfer volumes using past density data
        #.............................

        if(t == 0): # Initialization, transferring from 0h conical tubes to first competition flasks
            transferVolume_adjusted = 1.000

        if(t == transferTimeInterval): # First transfer

            N_WTctrl_thisInterval_init  = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+'WTctrl')&(allCompetitionData_growthCurves['time']==(t-transferTimeInterval+0.01)), 'density'].values[0]
            N_WTctrl_thisInterval_30mAgo = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+'WTctrl')&(allCompetitionData_growthCurves['time']==t-0.5), 'density'].values[0]

            N_2C3_thisInterval_init  = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+'2C-3')&(allCompetitionData_growthCurves['time']==(t-transferTimeInterval+0.01)), 'density'].values[0]
            N_2C3_thisInterval_30mAgo = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+'2C-3')&(allCompetitionData_growthCurves['time']==t-0.5), 'density'].values[0]

            N_comp_thisInterval_init    = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+competition)&(allCompetitionData_growthCurves['time']==(t-transferTimeInterval+0.01)), 'density'].values[0]

            r_WTctrl_thisInterval = (1/(transferTimeInterval-0.5))*numpy.log(N_WTctrl_thisInterval_30mAgo/N_WTctrl_thisInterval_init)
            r_2C3_thisInterval = (1/(transferTimeInterval-0.5))*numpy.log(N_2C3_thisInterval_30mAgo/N_2C3_thisInterval_init)

            N_comp_thisInterval_final_projected = N_comp_thisInterval_init * numpy.exp( r_2C3_thisInterval * transferTimeInterval )

            N_comp_nextInterval_init_projected = N_comp_thisInterval_final_projected * (transferVolume_adjusted/competitionFlaskMediaVolume)
            N_comp_nextInterval_final_projected_minRate = N_comp_nextInterval_init_projected * numpy.exp( r_2C3_thisInterval * transferTimeInterval )
            N_comp_nextInterval_final_projected_maxRate = N_comp_nextInterval_init_projected * numpy.exp( r_WTctrl_thisInterval * transferTimeInterval )

            if(N_comp_nextInterval_final_projected_minRate < min(transferIntervalEndDensityTargetRange) or N_comp_nextInterval_final_projected_maxRate > max(transferIntervalEndDensityTargetRange) ):
                N_comp_nextInterval_init_backCalc = max( min(transferIntervalInitDensityTargetRange), max(transferIntervalEndDensityTargetRange) / numpy.exp(r_WTctrl_thisInterval * transferTimeInterval) )

                transferVolume_backCalc = N_comp_nextInterval_init_backCalc/N_comp_thisInterval_final_projected * competitionFlaskMediaVolume
                transferVolume_adjusted = round( transferVolume_backCalc, 3)

        elif(t > transferTimeInterval): # Subsequent transfers
            N_WTctrl_thisInterval_init  = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+'WTctrl')&(allCompetitionData_growthCurves['time']==(t-transferTimeInterval+0.01)), 'density'].values[0]
            N_WTctrl_thisInterval_30mAgo = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+'WTctrl')&(allCompetitionData_growthCurves['time']==t-0.5), 'density'].values[0]

            N_comp_prevInterval_init    = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+competition)&(allCompetitionData_growthCurves['time']==(t-2*transferTimeInterval+0.01)), 'density'].values[0]
            N_comp_prevInterval_final    = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+competition)&(allCompetitionData_growthCurves['time']==(t-transferTimeInterval+0.00)), 'density'].values[0]
            N_comp_thisInterval_init    = allCompetitionData_growthCurves.loc[(allCompetitionData_growthCurves['culture']=='Competition '+competition)&(allCompetitionData_growthCurves['time']==(t-transferTimeInterval+0.01)), 'density'].values[0]

            r_WTctrl_thisInterval = (1/(transferTimeInterval-0.5))*numpy.log(N_WTctrl_thisInterval_30mAgo/N_WTctrl_thisInterval_init)
            r_comp_prevInterval = (1/(transferTimeInterval))*numpy.log(N_comp_prevInterval_final/N_comp_prevInterval_init)

            transferCalibrationReferenceRates.append(r_WTctrl_thisInterval)
            transferCalibrationReferenceRates.append(r_comp_prevInterval)

            N_comp_thisInterval_final_projected = N_comp_thisInterval_init * numpy.exp( r_WTctrl_thisInterval * transferTimeInterval )

            N_comp_nextInterval_init_projected = N_comp_thisInterval_final_projected * (transferVolume_adjusted/competitionFlaskMediaVolume)
            N_comp_nextInterval_final_projected_minRate = N_comp_nextInterval_init_projected * numpy.exp( min(transferCalibrationReferenceRates) * transferTimeInterval )
            N_comp_nextInterval_final_projected_maxRate = N_comp_nextInterval_init_projected * numpy.exp( max(transferCalibrationReferenceRates) * transferTimeInterval )

            if(N_comp_nextInterval_final_projected_minRate < min(transferIntervalEndDensityTargetRange) or N_comp_nextInterval_final_projected_maxRate > max(transferIntervalEndDensityTargetRange) ):
                N_comp_nextInterval_init_backCalc = max( min(transferIntervalInitDensityTargetRange), min(transferIntervalEndDensityTargetRange) / numpy.exp(min(transferCalibrationReferenceRates) * transferTimeInterval) )

                transferVolume_backCalc = N_comp_nextInterval_init_backCalc/N_comp_thisInterval_final_projected * competitionFlaskMediaVolume
                transferVolume_adjusted = round( transferVolume_backCalc, 3)

        #.............................
        #.............................

        print "\t\t\t\tTransfering flasks (transfer volume = " + str(transferVolume_adjusted*1000) + " uL)"
        competitionFlask_next = competitionFlaskTransfer(competitionFlask, 
                                                     media=LBStock, media_volume=competitionFlaskMediaVolume, 
                                                     transfer_volume=transferVolume_adjusted, 
                                                     name=competition)
        
        transferVolumes[competition].append(transferVolume_adjusted)


        competitionTimepointData = {'allele_freqs':[], 'pop_sizes':[]}

        
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~
        # Sample Competition Culture for Population Size Estimation (via flow cytometry enumeration):
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~
        print "\t\t\t\tPop size sampling"
        if(t==0):
            collectEnumerationSampleData(competition_culture=competitionFlask_next, 
                                         sample_reps=3, 
                                         t=t, 
                                         transfer_volumes=transferVolumes[competition],
                                         data=competitionTimepointData,
                                         sim_flow_cytometry=False)
        else:
            collectEnumerationSampleData(competition_culture=competitionFlask, 
                                         sample_reps=3, 
                                         t=t, 
                                         transfer_volumes=transferVolumes[competition],
                                         data=competitionTimepointData,
                                         sim_flow_cytometry=False)
        # Annotate this time point data with the competition name and rep number:
        competitionTimepointData['pop_sizes']    = [dict(d, competition=competition, competition_rep=competitionRep) for d in competitionTimepointData['pop_sizes']]
        # Add this competition's data recorded for this time point to the overall dataset:
        allCompetitionData_popSizes    = allCompetitionData_popSizes.append(competitionTimepointData['pop_sizes'])


        #~~~~~~~~~~~~~~~~~~~~~~~~~~~
        # Sample Competition for Allele Frequency Estimation (via sequencing):
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~
        print "\t\t\t\tAllele freq sampling"
        collectSequencingSampleData(competition_culture=competitionFlask, 
                                    sample_reps=1, 
                                    t=t, 
                                    data=competitionTimepointData)
        # Annotate this time point data with the competition name and rep number:
        competitionTimepointData['allele_freqs'] = [dict(d, competition=competition, competition_rep=competitionRep) for d in competitionTimepointData['allele_freqs']]
        # Add this competition's data recorded for this time point to the overall dataset:
        allCompetitionData_alleleFreqs = allCompetitionData_alleleFreqs.append(competitionTimepointData['allele_freqs'])

        
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~
        # Update the active flasks:
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~
        competitionFlasks[competition] = competitionFlask_next
        

    #~~~~~~~~~~~~~~~~~~~~~~~~~~~
    # Incubate the new  Competition Flasks:
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~
    print "\t\t\tIncubating flasks..."
    growthCurves = incubator.incubate(cultures=competitionFlasks.values(), time=transferTimeInterval, dt=0.01, record_growth_curves=True)
    growthCurves['time'] += t
    growthCurves['competition_rep'] = competitionRep
    allCompetitionData_growthCurves = allCompetitionData_growthCurves.append( growthCurves )
    

    if(t != 0):
        transferCount += 1


    # Update the time point
    t = t+transferTimeInterval
```

## Calculate Information & Load from Competition Data

```python
def KLDivergence(distP, distQ):
    distP = numpy.asarray(distP)
    distQ = numpy.asarray(distQ)
    distP[distP == 0] = numpy.finfo(float).eps
    distQ[distQ == 0] = numpy.finfo(float).eps
    print str( -1 * numpy.sum( distP * numpy.log2( distQ / distP ) ) )
    return -1 * numpy.sum( distP * numpy.log2( distQ / distP ) )

     
data_alleleFreqs = allCompetitionData_alleleFreqs

data_popSizes = allCompetitionData_popSizes
# These lines convert strings (such as marking 'missing data' in the spreadsheet) to NaN so they'll be excluded from calculations
data_popSizes['pop_size'] = pandas.to_numeric(data_popSizes['pop_size'], errors='coerce') # coerce converts strings to NaN
data_popSizes['density'] = pandas.to_numeric(data_popSizes['density'], errors='coerce') # coerce converts strings to NaN
# These lines enforce casting data vals to floats
data_popSizes = data_popSizes.astype({"pop_size": float})
data_popSizes = data_popSizes.astype({"density": float})
# Exclude -3 dilution samples 
data_popSizes = data_popSizes[data_popSizes.dilution != -3]


data_infoload = pandas.DataFrame( columns=['competition', 'competition_rep', 'time', 
                                            'infogain', 'freq_optimal_allele', 'infogain_asymptote',
                                            'subload', 'growth_rate_optimal_allele', 'growth_rate_population',
                                            'subload-infogain'] )

optimalAllele = 'WT'


for competition in data_alleleFreqs['competition'].unique():

    print "Competition " + competition + ":"
    for competition_rep in data_alleleFreqs['competition_rep'].unique():
        print "\trep " + str(competition_rep)


        data_alleleFreqs_compRep    = data_alleleFreqs[ (data_alleleFreqs['competition'] == competition) & (data_alleleFreqs['competition_rep'] == competition_rep) ]
        data_popSizes_compRep       = data_popSizes[ (data_popSizes['competition'] == competition) & (data_popSizes['competition_rep'] == competition_rep) ]


        initAlleleFreqDist      = None
        initFreqOptimalAllele   = None
        infogainAsymptote       = None

        popSizes_previous       = None
        t_previous              = None

        growthRate_optimalAllele_t  = 0.0
        growthRate_population_t     = 0.0

        infogain_t              = 0.0
        subload_t               = 0.0

        for t in data_alleleFreqs_compRep['time'].unique():

            data_alleleFreqs_compRep_t = data_alleleFreqs_compRep[ (data_alleleFreqs_compRep['time'] == t) ]
            data_popSizes_compRep_t = data_popSizes_compRep[ (data_popSizes_compRep['time'] == t) ]

            alleleFreqs_t = sorted( zip(data_alleleFreqs_compRep_t['allele'], data_alleleFreqs_compRep_t['allele_freq']), key=lambda tup: tup[0] ) 

            freqOptimalAllele_t = dict(alleleFreqs_t)[optimalAllele]

            alleleFreqDist_t = [alleleFreq[1] for alleleFreq in alleleFreqs_t]

            popSizes_t = data_popSizes_compRep_t.groupby('population', as_index=False)['pop_size'].mean()
            popSizes_t = dict( zip(popSizes_t['population'], popSizes_t['pop_size']) )

            densities_t = data_popSizes_compRep_t.groupby('population', as_index=False)['density'].mean()
            densities_t = dict( zip(densities_t['population'], densities_t['density']) )

            if(t == 0):
                initAlleleFreqDist = alleleFreqDist_t
                initFreqOptimalAllele = freqOptimalAllele_t
                infogainAsymptote = -1*numpy.log2(initFreqOptimalAllele)

                infogain_t  = 0.0
                subload_t   = 0.0

            else:

                infogain_t  = KLDivergence(alleleFreqDist_t, initAlleleFreqDist)

                deltaT = t - t_previous

                growthRate_optimalAllele_t = (1/deltaT) * numpy.log2(popSizes_t['gfp'] / popSizes_previous['gfp'])

                growthRate_population_t = (1/deltaT) * numpy.log2(popSizes_t['total'] / popSizes_previous['total'])

                subload_t   = subload_t + (growthRate_optimalAllele_t - growthRate_population_t) * deltaT
                
                print (1/deltaT)
                print 


            data_infoload = data_infoload.append( [{'competition':competition, 'competition_rep':competition_rep, 'time':t, 
                                                    'infogain':infogain_t, 'freq_optimal_allele':freqOptimalAllele_t, 'infogain_asymptote':infogainAsymptote,
                                                    'subload': subload_t, 'growth_rate_optimal_allele':growthRate_optimalAllele_t, 'growth_rate_population':growthRate_population_t,
                                                    'subload-infogain':(subload_t - infogain_t)
                                                    }] )

            
            t_previous          = t
            popSizes_previous   = popSizes_t





```
